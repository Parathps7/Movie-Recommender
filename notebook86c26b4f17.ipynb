```python
import ast
# Now we gonna make only 3 rows
# MovieID Title Tags(this contains rest of all columns)
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
# To convert string to python object as listed in string
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')

# check
```

```python
def convert5(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 5:
            L.append(i['name'])
        counter+=1
    return L 
# we keep only first 5 actors from cast
```

```python
movies['cast'] = movies['cast'].apply(convert5)
movies.head()
```

```python
# movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director' or i['job'] == 'Writer' or i['job'] == 'Screenplay':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
# removed space between words
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()

```

```python
import nltk
new.head()
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
nltk.download('stopwords')
ps=PorterStemmer()
new.head()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        if i.lower() not in set(stopwords.words('english')):
            y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags']=new['tags'].apply(stem)
new.head()
```

```python
# Now we will try to convert tags to vector(2d space)(math wala) to get similarities within movies.If we try to compare maximum number of words for realation that will be bad idea,as it is not an ideal way

# We then make a dataframe with movies and 5000 most common words and plot them in 5000 dimensional space,the closest vectors will be the recommended movies
# We won't consider stop words like or,and of
# we won't calculate the euclidean distance but cosine distance bewtween them i.e angle.

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
cv.get_feature_names_out()
```

```python
# ['love','loved','loving' ] converts To 
# ['love','love','love']
# this is called strmming,now we'll do it
# we will use nltk library,i.e a natural language processing library
```

```python
# Calculating cosine distance
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity(vector).shape
```

```python
similarity = cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
# sort acc to second  part
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    result=[]
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        result.append(new.iloc[i[0]].title)
    return result
        
    # we get the movie index and find similarity[index] and then sort and get the top 5 similar movies
```

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

# Function to add two numbers (replace this with your actual function from the Jupyter Notebook)

@app.route('/recommend/<string:movie_name>', methods=['GET'])
def movie_recommender(movie_name):
    if movie_name is None:
        return jsonify(error="Invalid request. Please provide a valid 'movie_name'."), 400

    result = recommend(movie_name)
    print(result)
    return jsonify(result)


if __name__ == '__main__':
    app.run()
# To make it into an API using flask
```
